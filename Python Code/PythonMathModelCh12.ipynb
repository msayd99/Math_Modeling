## Chapter 12: Concepts of Machine Learning
### K-means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster
import warnings
```

```python
warnings.filterwarnings('ignore')
mydata = np.random.uniform(size=(250,2))

kmeans = cluster.KMeans(n_clusters=5, random_state=0)
y_mean = kmeans.fit_predict(mydata)
centers = kmeans.cluster_centers_

colors = ['black', 'red', 'green', 'blue', 'purple']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,12))

for (i, cent, colour) in zip(range(len(colors)), centers, colors):
    ax.plot(cent[0], cent[1], color=colour, marker='*', ms=12)
    ax.plot(mydata[y_mean==i,0], mydata[y_mean==i,1], color=colour,
            marker='o', ms=8, ls='', label=f"Cluster {i+1}")

ax.set_ylim(-0.05, 1.1)
ax.set_xlabel("$x_1$",size=20)
ax.set_ylabel("$x_2$",size=20)
ax.tick_params(labelsize=20)
ax.legend(loc='upper center', ncol=5)
ax.grid()

ax.set_title("K-means Clustering of 250 Uniformly \nRandom Points"
             ,fontweight="bold",size=25)

#fig.savefig("K-MeansClustering.jpg", bbox_inches='tight')
```
